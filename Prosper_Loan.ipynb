#### Target Column showing as multiple values but we need to convert them into Binary Classification

`Chargedoff`: - A loan is considered â€œcharged-offâ€ after a borrower misses 5 consecutive monthly payments. At that point, the loan is considered unlikely to be repaid, We can considered as loan application to be `REJECTED` here.

`Current`: - Current Loan means, as of any date, a Loan as to which all payments of principal and interest then due have been paid, except for the most recent payment then due of principal and interest. We can considered as loan application to be `ACCEPTED` here.

`Completed`: - When a Lender has transferred funds across to the borrowing party they will mark the Loan Enquiry as Completed.We can considered as loan application to be `ACCEPTED` here.

`Defaulted`: - Default is the failure to make required interest or principal repayments on a debt, whether that debt is a loan or a security. We can considered as loan application to be `REJECTED` here.

`FinalPaymentInProgress`: - The interest and the principal is already paid but final approval or loan completion document is in progress.We can considered as loan application to be `ACCEPTED` here.

`Cancelled` : We should considered application to be `REJECTED` here as might be application was cancelled due to any froad and improper document.

`For Past Due`, we can consider `ACCEPTED` till 120 days and after 120 days it would be consider as `REJECTED`.

````python
mapping = {
    "Chargedoff":0,
    "Defaulted":0,
    "Cancelled":0,
    "Past Due (>120 days)":0,
    "Current":1,
    "Completed":1,
    "FinalPaymentInProgress":1,
    "Past Due (1-15 days)":1,
    "Past Due (31-60 days)":1,
    "Past Due (61-90 days)":1,
    "Past Due (91-120 days)":1,
    "Past Due (16-30 days)":1

}

data["LoanStatus"] = data["LoanStatus"].map(mapping)
````
